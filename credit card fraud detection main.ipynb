```python
#distribution of legit and fraud transaction 
credit_card_df['Class'].value_counts()
```

```python
#as we can see this is highly undistributed dataset so for better accuracy we can take random 492 rows from legit transcation and concatenate them with fraud one to get a good distributed dataset
```

```python
legit=credit_card_df[credit_card_df.Class==0]
```

```python
fraud=credit_card_df[credit_card_df.Class==1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#duing undersampling to build a dataset with coantaning similar distribution of fraud and legit transaction
legit_sample=legit.sample(n=492)
```

```python
#concating both dataset to build a new one
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#splitting dataset into features andb target
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,stratify=Y,random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```
